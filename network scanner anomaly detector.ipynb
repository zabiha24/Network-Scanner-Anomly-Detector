Step 1 - Cleaning, Loading, Preprocessing, and Exploring the NSL-KDD dataset.

```python
import pandas as pd

# Define the column names for the NSL-KDD dataset
columns = [
    "duration", "protocol_type", "service", "flag", "src_bytes", "dst_bytes", "land", "wrong_fragment",
    "urgent", "hot", "num_failed_logins", "logged_in", "num_compromised", "root_shell", "su_attempted",
    "num_root", "num_file_creations", "num_shells", "num_access_files", "num_outbound_cmds",
    "is_host_login", "is_guest_login", "count", "srv_count", "serror_rate", "srv_serror_rate",
    "rerror_rate", "srv_rerror_rate", "same_srv_rate", "diff_srv_rate", "srv_diff_host_rate",
    "dst_host_count", "dst_host_srv_count", "dst_host_same_srv_rate", "dst_host_diff_srv_rate",
    "dst_host_same_src_port_rate", "dst_host_srv_diff_host_rate", "dst_host_serror_rate",
    "dst_host_srv_serror_rate", "dst_host_rerror_rate", "dst_host_srv_rerror_rate", "label", "difficulty"
]

# Load the training dataset
train_data_path = 'E:/University/Sem 3/AI/Project/NSL-KDD-Dataset/KDDTrain+.csv'
df_train = pd.read_csv(train_data_path, names=columns, header=None)

# Load the test dataset
test_data_path = 'E:/University/Sem 3/AI/Project/NSL-KDD-Dataset/KDDTrain+.csv'
df_test = pd.read_csv(test_data_path, names=columns, header=None)

# Display the first few rows of the training dataset
print("Training Dataset:")
display(df_train.head())

# Display the first few rows of the test dataset
print("Test Dataset:")
display(df_test.head())
```

```python
# Display the first few rows of the training dataset
print("Training Dataset Head:")
print(df_train.head())

# Summary of the training dataset
print("Training Dataset Info:")
print(df_train.info())
print("Training Dataset Description:")
print(df_train.describe())
```

```python
# Check for null values in the training dataset
print("Null values in Training Dataset:")
print(df_train.isnull().sum())

# Check for null values in the test dataset
print("Null values in Test Dataset:")
print(df_test.isnull().sum())
```

```python
from sklearn.preprocessing import LabelEncoder
# Label encoding for categorical features
categorical_features = ["protocol_type", "service", "flag"]
encoder = LabelEncoder()

for feature in categorical_features:
    df_train[feature] = encoder.fit_transform(df_train[feature])
    df_test[feature] = encoder.transform(df_test[feature])

# Display the first few rows of the training dataset to verify the changes
print("Training Dataset after Label Encoding:")
display(df_train.head())

# Display the first few rows of the test dataset to verify the changes
print("Test Dataset after Label Encoding:")
display(df_test.head())
```

```python
from sklearn.preprocessing import StandardScaler
# Define features and target variable for training dataset
X_train = df_train.drop("label", axis=1)
y_train = df_train["label"]D

# Define features and target variable for test dataset
X_test = df_test.drop("label", axis=1)
y_test = df_test["label"]

# Standardize numeric features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Display the first few rows of the scaled training features
print("Scaled Training Features:")
print(pd.DataFrame(X_train_scaled, columns=X_train.columns).head())

# Display the first few rows of the scaled test features
print("Scaled Test Features:")
print(pd.DataFrame(X_test_scaled, columns=X_test.columns).head())
```

```python
from sklearn.preprocessing import LabelEncoder

# Convert labels to binary
df_train['label'] = df_train['label'].apply(lambda x: 'normal' if x == 'normal' else 'attack')
df_test['label'] = df_test['label'].apply(lambda x: 'normal' if x == 'normal' else 'attack')

# Initialize label encoder
encoder = LabelEncoder()

# Apply label encoding to the binary labels for training dataset
y_train = encoder.fit_transform(df_train['label'])

# Apply label encoding to the binary labels for test dataset
y_test = encoder.transform(df_test['label'])

# Display the first few rows of the encoded labels
print("Encoded Labels for Training Data:")
print(y_train[:5])

print("Encoded Labels for Test Data:")
print(y_test[:5])
```

```python
from sklearn.model_selection import train_test_split
# Assuming X_train_scaled and y_train are already defined
X_train_final, X_val, y_train_final, y_val = train_test_split(X_train_scaled, y_train, test_size=0.2, random_state=42)

# Display the shapes of the resulting datasets
print("X_train_final shape:", X_train_final.shape)
print("X_val shape:", X_val.shape)
print("y_train_final shape:", y_train_final.shape)
print("y_val shape:", y_val.shape)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix

# Train a simple model
model = LogisticRegression()
model.fit(X_train_scaled, y_train)

# Make predictions
y_pred = model.predict(X_test_scaled)

# Evaluate the model
print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred))
print("\nClassification Report:")
print(classification_report(y_test, y_pred))
```

```python
from scapy.all import *
from datetime import datetime

# Define the threshold for categorizing packets as anomaly
PACKET_SIZE_THRESHOLD = 1500  # Adjust this value as needed

def packet_callback(packet):
    if packet.haslayer(IP):
        ip_src = packet[IP].src
        ip_dst = packet[IP].dst
        if packet.haslayer(TCP):
            protocol = "TCP"
            sport = packet[TCP].sport
            dport = packet[TCP].dport
        elif packet.haslayer(UDP):
            protocol = "UDP"
            sport = packet[UDP].sport
            dport = packet[UDP].dport
        else:
            protocol = "Other"
            sport = None
            dport = None

        # Determine packet size
        packet_size = len(packet)

        # Categorize packet as Normal or Anomaly based on size
        if packet_size > PACKET_SIZE_THRESHOLD:
            category = "Anomaly"
        else:
            category = "Normal"

        # Get current date and time
        current_time = datetime.now().strftime("%Y-%m-%d %H:%M:%S")

        # Print packet details along with the category and current time
        print(f"Date & Time: {current_time},      IP: {ip_src} -> {ip_dst},      Protocol: {protocol},      Src Port: {sport},      Dst Port: {dport},      Size: {packet_size},      Category: {category}")

# Sniff network traffic and call packet_callback for each packet
sniff(prn=packet_callback, filter="ip", store=0)
```
